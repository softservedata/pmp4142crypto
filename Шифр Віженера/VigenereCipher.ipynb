Шифр Віженера  — поліалфавітний шифр, який як ключ використовує слово. 

В наступній таблиці горизонтальна лінія - вхідний текст, вертикальна - ключ


#### шифрування

```python
def VigenereEncode(plaintextFromUser, keyFromUser):
    # Ініціалізуємо алфавіт
    alphabet = "абвгґдеєжзиіїйклмнопрстуфхцчшщьюя"
    input_string = ""
    enc_key = ""
    enc_string = ""

    # Ключ
    enc_key = keyFromUser
    # Приводимо всі букви до малого регістру
    enc_key = enc_key.lower()

    # Вхідний текст
    input_string = plaintextFromUser
    # Приводимо всі букви до малого регістру
    input_string = input_string.lower()

    string_length = len(input_string)

    # розширюємо ключ до розміру вхідного тексту
    expanded_key = enc_key
    expanded_key_length = len(expanded_key)

    while expanded_key_length < string_length:
        expanded_key = expanded_key + enc_key
        expanded_key_length = len(expanded_key)

    key_position = 0

    for letter in input_string:
        if letter in alphabet:
            # Знаходимо позицію кожної букви в алфавіті
            position = alphabet.find(letter)
            key_character = expanded_key[key_position]
            key_character_position = alphabet.find(key_character)
            key_position = key_position + 1
            # Задаємо нову позицію букві
            new_position = position + key_character_position
            if new_position > 33:
                new_position = new_position - 33
            new_character = alphabet[new_position]
            enc_string = enc_string + new_character
        else:
            enc_string = enc_string + letter
    return(enc_string)
```

#### розшифрування

```python
def VigenereDecode(plaintextFromUser, keyFromUser):
    alphabet = "абвгґдеєжзиіїйклмнопрстуфхцчшщьюя"
    input_string = ""
    decKey = ""
    decString = ""
    decKey = keyFromUser
    decKey = decKey.lower()
    input_string = plaintextFromUser
    input_string = input_string.lower()

    string_length = len(input_string)

    expanded_key = decKey
    expanded_key_length = len(expanded_key)

    while expanded_key_length < string_length:
        expanded_key = expanded_key + decKey
        expanded_key_length = len(expanded_key)

    key_position = 0

    for letter in input_string:
        if letter in alphabet:
            position = alphabet.find(letter)
            key_character = expanded_key[key_position]
            key_character_position = alphabet.find(key_character)
            key_position = key_position + 1
            new_position = position - key_character_position
            if new_position > 33:
                new_position = new_position + 33
            new_character = alphabet[new_position]
            decString = decString + new_character
        else:
            decString = decString + letter
    return(decString)
```
