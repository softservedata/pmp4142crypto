Після визначення висоти частокола додати у вихідний текст надлишкову інформацію. Дані читати та записувати у файли.

```python
# імпортуємо бібліотеку math для квадратного кореня
import math
```

### Шифрування

```python
def encryptRailFence(text, height):
  
    # Створимо матрицю для шифрування
    # plain text height = рядки 
    # length(text) = колонки
    rail = [['\n' for i in range(len(text))]
                  for j in range(height)]
      
    dir_down = False
    row, col = 0, 0
      
    for i in range(len(text)):
          
        if (row == 0) or (row == height - 1):
            dir_down = not dir_down
          
        rail[row][col] = text[i]
        col += 1
          
        if dir_down:
            row += 1
        else:
            row -= 1
    result = []
    for i in range(height):
        for j in range(len(text)):
            if rail[i][j] != '\n':
                result.append(rail[i][j])
    return("" . join(result))
```

### Розшифрування

```python
def decryptRailFence(encodedText, height):

    rail = [['\n' for i in range(len(encodedText))] 
                  for j in range(height)]
      
    dir_down = None
    row, col = 0, 0
      
    for i in range(len(encodedText)):
        if row == 0:
            dir_down = True
        if row == height - 1:
            dir_down = False
          
        rail[row][col] = '*'
        col += 1
          
        if dir_down:
            row += 1
        else:
            row -= 1
              
    index = 0
    for i in range(height):
        for j in range(len(encodedText)):
            if ((rail[i][j] == '*') and
               (index < len(encodedText))):
                rail[i][j] = encodedText[index]
                index += 1
          
    result = []
    row, col = 0, 0
    for i in range(len(encodedText)):     
        if row == 0:
            dir_down = True
        if row == height-1:
            dir_down = False
              
        if (rail[row][col] != '*'):
            result.append(rail[row][col])
            col += 1
            
        if dir_down:
            row += 1
        else:
            row -= 1
    return("".join(result))
```

### Тестуємо програму

```python
# дістаємо висоту за умовою як квадратний корінь з кількості символів
height = int(math.sqrt(len("Частокол!")))
# Якщо є потреба зчитати з файлу розкоментуйте
#f = open("railFenceExample.txt", "r")
#text = f.readline()

print("Шифрований текст")
print(encryptRailFence("Частокол!", height))
 
print("Розшифрований текст")
print(decryptRailFence("Чо!атклсо", height))
```
